```python
import numpy as np

arr_zeros = np.zeros(5)
print(" Zeros Array  : ",arr_zeros)

print("")

arr_ones=np.ones(8)
print(" Ones Array : ",arr_ones)

print("")

arr_1_20 = np.arange(1,21)
print(" 1 to 20 : ",arr_1_20)

print("")

arr_even = np.arange(2,20,2)
print(" Even Numbers : ",arr_even)

print("")

linspace_arr = np.linspace(0,1,10)
print(" Equally Spaced Values : ",linspace_arr)
```
