```python
import numpy as np

Matrix = np.array ([[2,4,6],
          [8,10,12],
          [14,16,18]])

print(" Shape of Matrix : ",Matrix.shape)

print("")

print(" Row 2 Column 3 : ",Matrix[1,2])

print("")

print(" First Row : ",Matrix[0])

print("")

print(" Second Column : ",Matrix[:,1])

print("")

print(" Sum Of All Elements : ",np.sum(Matrix))
```
