```python
import pandas as pd

Data = {
    "Name":["Ali","Sara","John","Ayesha"],
    "Age":[20,21,19,22],
    "Marks":[85,90,78,88]
}

df = pd.DataFrame(Data)

print("Average Marks : ",df["Marks"].mean())

print("")

print("Highest Marks : ",df["Marks"].max())

print("")

print("Lowest Marks : ",df["Marks"].min())
```
