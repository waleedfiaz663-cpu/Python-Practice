```python
import numpy as np

Data = np.array([12,15,18,20,22,25,30])

print("Mean : ",np.mean(Data))

print("")

print("Median : ",np.median(Data))

print("")

print("Maximum Value : ",np.max(Data))

print("")

print("Minimum Value : ",np.min(Data))

print("")

print("Sum of all values : ",np.sum(Data))
```
