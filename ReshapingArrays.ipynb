```python
import numpy as np

arr = np.arange(1,13)

matrix_3x4 = np.reshape(arr,(3,4))

matrix_4x3 = np.reshape(arr,(4,3))

print("Original Array : ")

print("")

print(arr)

print("")

print("3x4 Matrix : ")

print("")

print(matrix_3x4)

print("")

print("4x3 Mattrix : ")

print("")

print(matrix_4x3)
```
