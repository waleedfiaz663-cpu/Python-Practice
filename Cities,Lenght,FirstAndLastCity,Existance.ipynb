```python
cities=("Abbottabad","Islamabad","Lahore","Karachi","Peshwar")

total_len=len(cities)

print("total len :",total_len)

first_city=cities[0]
last_city=cities[-1]

print("first city :",first_city)
print("last city :",last_city)

city=input("Please Enter City :")

if city.title() in cities:

     print("city Exist in the Tuple")

else:

     print("city Does not Exist in tuple")
```
