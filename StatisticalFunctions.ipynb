```python
import numpy as np

Marks = np.array([78,85,92,67,88,95,73,81])

print(" Maximum value : ",np.max(Marks))

print("")

print(" Minimum Marks : ",np.min(Marks))

print("")

print(" Mean : ",np.mean(Marks))

print("")

print(" Median : ",np.median(Marks))

print("")

print(" Standard Deviation : ",np.std(Marks))
```
