```python
import numpy as np

Data = np.array([12,15,20,25,30,35])

print("Mean is :",np.mean(Data))

print("")

print("Median is : ",np.median(Data))

print("")

print("Standard Deviation is :",np.std(Data))

print("")

print("Maximum Value is : ",np.max(Data))

print("")

print("Minimum Value is : ",np.min(Data))
```
