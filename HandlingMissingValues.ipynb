```python
import pandas as pd

Data = {
    "Name":["Ali","Sara","John",None],
    "Marks":[85,None,78,88]
}

df=pd.DataFrame(Data)

print(" Finding Missing Values : ")

print("")

print(df.isnull())

df["Name"] = df["Name"].fillna("Waleed")

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print("")

print("Cleaned DataFrame : ")

print("")

print(df)

```
