```python
import numpy as np

Marks = np.array([78,85,92,67,88,95,73,81,90,76])

Average = np.mean(Marks)

print(" Average Marks : ",Average)

print("")

print(" Maximum Marks : ",np.max(Marks))

print("")

print(" Minimum Marks : ",np.min(Marks))

print("")

print("====== Marks Greater Than Average ======")

print("")

print(Marks[Marks>Average])
```
