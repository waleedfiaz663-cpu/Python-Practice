```python
import pandas as pd

Data = {
    "Name":["Ali","Sara","John","Ayesha"],
    "Age":[20,21,19,22],
    "Marks":[85,90,78,88]
}

df=pd.DataFrame(Data)

print("Name and Marks : ")

print("")

print(df[["Name","Marks"]])

print("")

print("Marks Greater Than 85 : ")

print("")

print(df[df["Marks"]>85])
```
