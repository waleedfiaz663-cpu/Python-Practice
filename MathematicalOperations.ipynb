```python
import numpy as np

A = np.array([1,2,3,4,5])

B = np.array([5,4,3,2,1])

print("Addition : ", A+B)

print("")

print("Subtraction : ",A-B)

print("")

print("Multiplication : ",A*B)

print("")

print("Division : ",A/B)

print("")

print("Square of A : ",A**2)
```
