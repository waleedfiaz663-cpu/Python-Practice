```python
import numpy as np

arr =([5,10,15,20,25,30,35,40])

print(" First Element : ",arr[0])

print("")

print(" last Element : ",arr[-1])

print("")

print(" Index 2 to 5 : ",arr[2:6])

print("")

print(" First Four Elements : ",arr[:4])

print("")

print(" All Except First Two : ",arr[2:])
```
